# DFT+eDMFT

First, we need to optimize the lattice structure using WIEN2k. You can find the generated GaAs structure file in the GitHub repository.

Before running the DMFT code, we need to generate the supercell structure with some atoms displaced from their equilibrium positions.


```python
%%bash
phonopy -d --dim="2 2 2" -c GaAs.struct --wien2k
```

It generates 5 different supercell structures GaAs.structS, GaAs.structS-001, GaAs.structS-002, GaAs.structS-003, FeSe.structS-004. We should run the DMFT calculations for of these structures files. After that we can employ DMFT calculations for each of them.

Now we perform self consistent calculations inside each of the 4 directories, for each of the 4 structures. Here we just show the process for one of them.

```python
cd GaAs1
init_lapw
```
